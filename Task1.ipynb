### Create a bar chart or histogram to visualize the distribution of a categorial or continuous variable, such as the distribution of ages or genders in a population

Importing the necessary Libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Reading the data

```python
data = pd.read_csv("population.csv")
data.head(10)
```

Total Distinct Countries

```python
len(data['Country Name'].unique())
```

Distinct Countries 

```python
data['Country Name'].unique()
```

Bar Chart showing Population Distribution

```python
#List of contries for the graph.
countries_to_plot = ['Ukraine', 'Korea, Rep.', 'Saudi Arabia', 'Japan', 'Qatar', 'France','Bangladesh','Senegal']

#Filtering the data frame
filtered_data = data[data['Country Name'].isin(countries_to_plot)]

# Create bars
color = ['green', 'blue', 'purple', 'red','pink', 'black','yellow','orange']
fig, ax = plt.subplots()
ax.bar(filtered_data['Country Name'], filtered_data['2023'], color=color)

plt.xlabel ('Country Name')
plt.ylabel ('Population in 2023')
plt.title('Population of selected countries in 2023')
plt.xticks(rotation=45, ha='right') 
plt.show()
```
